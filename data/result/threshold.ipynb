```python
import pandas as pd
from sklearn.metrics import accuracy_score, precision_score, recall_score

df = pd.read_csv("result_wuhan_labeled_2class_ABiLSTM_Att_weibo300_JD.csv")
df.sort_values("predictions",inplace=True,ascending=False)
score = df["predictions"].tolist()
pre_class = df["pre_class"].tolist()
true_class =df["true_class"].tolist()

acc=[] 
pre=[]
rec=[]
for i in range(len(pre_class)):#每次读一行
    pre_class[i]=1
    acc.append(accuracy_score(true_class,pre_class))
    pre.append(precision_score(true_class, pre_class, average='weighted'))
    rec.append(recall_score(true_class, pre_class, average='weighted'))
```

```python
import matplotlib.pyplot as plt
plt.figure(dpi=300,figsize=(8,3))
plt.suptitle('评估结果', fontsize=14, y=1.1)
plt.rcParams['font.sans-serif']=['SimSun'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False   #用来正常显示负号
plt.rcParams['axes.titlepad'] = 10
plt.subplot(1,3,1)
plt.plot(score,acc)
plt.xlabel('score')
plt.ylabel('accuracy')
p_x=score[acc.index(max(acc))]
p_y=max(acc)
plt.scatter([p_x], [p_y], s=25, c='r')  # stroke, colour
plt.text(p_x+0.15, p_y+0.015, str(p_x)+','+str(p_y), ha='center', va='bottom', fontsize=7)  # horizontal alignmen
plt.subplot(1,3,2)
plt.plot(score,pre)
plt.xlabel('score')
plt.ylabel('precision')
p_x=score[pre.index(max(pre))]
p_y=max(pre)
plt.scatter([p_x], [p_y], s=25, c='r')  # stroke, colour
plt.text(p_x+0.15, p_y+0.015, str(p_x)+','+str(p_y), ha='center', va='bottom', fontsize=7)  # horizontal alignmen
plt.subplot(1,3,3)
plt.plot(score,rec)
plt.xlabel('score')
plt.ylabel('recall')
p_x=score[rec.index(max(rec))]
p_y=max(rec)
plt.scatter([p_x], [p_y], s=25, c='r')  # stroke, colour
plt.text(p_x+0.15, p_y+0.015, str(p_x)+','+str(p_y), ha='center', va='bottom', fontsize=7)  # horizontal alignmen
plt.tight_layout()
plt.show()
```
